```python
import random

# Path file .txt yang akan dibuat
text_file_path = "number.txt"

# Fungsi untuk menghasilkan angka random dengan tanda desimal
def generate_decimal_numbers(num_lines):
    lines = []
    for _ in range(num_lines):
        integer_part = random.randint(0, 999)  # Bagian angka sebelum desimal
        fractional_part = random.randint(0, 99)  # Bagian angka setelah desimal
        decimal_number = f"{integer_part}.{fractional_part:02d}"  # Format dengan 2 digit desimal
        lines.append(decimal_number)
    return lines

# Jumlah angka yang ingin dibuat
num_lines = 100  # Ganti sesuai kebutuhan

# Membuat angka random dengan tanda desimal
numbers = generate_decimal_numbers(num_lines)

# Menulis angka tersebut ke file .txt
with open(text_file_path, "w") as file:
    file.write("\n".join(numbers))

print(f"File '{text_file_path}' berhasil dibuat dengan {num_lines} angka random!")
```

```python
import random
from PIL import Image, ImageFont, ImageDraw, ImageFilter, ImageEnhance, ImageOps
import os

# Paths
font_path = "7segment.ttf"  # Path ke font seven segment
background_path = "bg.jpg"  # File background gambar
output_dir = "generated_images"  # Direktori output untuk menyimpan gambar
text_file_path = "number.txt"  # File output untuk menyimpan angka yang dihasilkan

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to generate a random word with 6 characters (5 digits + 1 decimal point)
def generate_random_word():
    integer_part = random.randint(0, 9)  # Bagian sebelum titik desimal (minimal 3 digit)
    fractional_part = random.randint(0, 99)  # Bagian setelah titik desimal (2 digit)
    return f"{integer_part}.{fractional_part:03d}"  # Formatkan menjadi angka desimal

# Function to apply a minimal blur effect
def apply_blur(image):
    return image.filter(ImageFilter.GaussianBlur(radius=random.uniform(1.0, 1.5)))  # Efek blur ringan

# Function to adjust brightness slightly darker
def adjust_brightness(image):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(random.uniform(0.7, 0.9))  # Kurangi brightness (1.0 = asli, <1.0 = lebih gelap)

# Function to stretch text vertically
def stretch_text_vertically(image, stretch_factor=1.5):
    width, height = image.size
    new_height = int(height * stretch_factor)  # Memperbesar tinggi
    return image.resize((width, new_height), Image.Resampling.LANCZOS)

# Function to apply italic effect by shearing the text
def apply_italic_effect(text_image, shear_factor=0.3):
    width, height = text_image.size
    m = shear_factor  # Positive shear to tilt to the right
    transform_matrix = (1, m, 0, 0, 1, 0)
    new_width = width + int(height * abs(m))
    text_image = text_image.transform((new_width, height), Image.AFFINE, transform_matrix, Image.BICUBIC)
    return text_image

# Buka background gambar
background = Image.open(background_path).convert("RGBA")

# Simpan angka yang dihasilkan ke dalam file
generated_numbers = []

# Generate 25 synthetic images
for i in range(25):
    # Generate angka dengan format yang diinginkan
    extended_word = generate_random_word()
    generated_numbers.append(extended_word)  # Simpan angka ke dalam daftar

    # Prepare font and text size
    font_size = int(min(background.width, background.height) * 0.7)
    font = ImageFont.truetype(font_path, font_size)
    draw = ImageDraw.Draw(background)
    text_width, text_height = draw.textbbox((0, 0), extended_word, font=font)[2:4]

    # Adjust font size if text is too large for the background
    while text_width > background.width or text_height > background.height:
        font_size -= 5
        font = ImageFont.truetype(font_path, font_size)
        text_width, text_height = draw.textbbox((0, 0), extended_word, font=font)[2:4]

    # Calculate the position to center the text on the background
    position = ((background.width - text_width) // 2, (background.height - text_height) // 2)

    # Adjust the vertical position to move the text upwards
    position = (position[0], position[1] - 20)  # Change 30 to any value you need to move it upwards

    word_img = Image.new("RGBA", background.size, (255, 255, 255, 0))
    word_draw = ImageDraw.Draw(word_img)
    word_draw.text(position, extended_word, fill=(0, 0, 0), font=font)

    # Apply italic effect
    word_img = apply_italic_effect(word_img)

    # Apply blur
    word_img = apply_blur(word_img)

    # Stretch the text vertically
    word_img = stretch_text_vertically(word_img, stretch_factor=1.5)

    # Combine with background
    combined_img = background.copy()
    combined_img.paste(word_img, (0, 0), word_img)

    # Apply brightness adjustment (slightly darker)
    combined_img = adjust_brightness(combined_img)

    # Replace invalid filename characters
    safe_filename = "".join(c if c.isalnum() or c in [' ', '.', ',', '-', "'"] else '_' for c in extended_word)

    # Save the image
    image_filename = os.path.join(output_dir, f"{safe_filename}.png")
    combined_img.save(image_filename)
    print(f"Saved: {image_filename}")

# Simpan angka yang dihasilkan ke dalam file teks
with open(text_file_path, "w") as file:
    file.write("\n".join(generated_numbers))

print(f"Generated 25 images in '{output_dir}' and saved numbers to '{text_file_path}'")
```

```python
import os
from openpyxl import Workbook

# Path ke direktori gambar
image_dir = "generated_images"  # Ganti dengan direktori gambar Anda
excel_file = "generated_images_labels.xlsx"  # Nama file Excel yang akan dibuat

# Dapatkan semua file gambar di dalam direktori
image_files = [f for f in os.listdir(image_dir) if f.endswith(".png")]

# Buat workbook dan sheet
wb = Workbook()
ws = wb.active
ws.title = "Image Labels"

# Tulis header
ws.append(["image", "label"])

# Tulis data ke Excel
for image_file in image_files:
    # Ambil nama file tanpa ekstensi (misalnya "940.53.png" menjadi "940.53")
    label = os.path.splitext(image_file)[0]
    
    # Menambahkan data ke Excel
    ws.append([image_file, label])

# Simpan file Excel
wb.save(excel_file)

print(f"Excel file '{excel_file}' has been created.")
```
