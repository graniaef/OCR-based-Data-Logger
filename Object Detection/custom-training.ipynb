```python
# Create .yaml file 
import yaml

data_yaml = dict(
    train = '/kaggle/tmp/sample-5/train',
    val = '/kaggle/tmp/sample-5/valid',
    test = '/kaggle/tmp/sample-5/test',
    nc = 1,
    names = ['display'])

with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
```

```python
from ultralytics import YOLO

# Load model YOLOv8s
model = YOLO('yolov8s.pt')

# Latih model
model.train(data='data.yaml', epochs=50, imgsz=640, batch=32)
```
